# Find Thresholds for Similarity

```python
import numpy as np
```

```python
#https://www.sbert.net/docs/sentence_transformer/usage/semantic_textual_similarity.html#similarity-calculation

from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
```

Roles

```python
#control group, similarity expected to be 1 
same_roles = [
    ("customer","customer"), # Computer Repair Service
    ("CRS","CRS"), # Computer Repair Service
    ("employee","employee"), # Expense Handling Process
    ("a manager","a manager"), # Expense Handling Process
    ("Puori","Puori"), # Quality Control of medical products
    ("Ellipse","Ellipse"), # Quality Control of medical products
    ("a customer","a customer"), # Mikkeller webshop
    ("The payment method used in the billing details","The payment method used in the billing details"), # Mikkeller webshop
    ("Swapfiets store employees","Swapfiets store employees"), # Bicycle Rental Process
    ("the staff","the staff"), # Bicycle Rental Process
    ("the seller","the seller"), # Long term renting process
    ("management","management"), # Long term renting process
    ("P&R","P&R"), # Design Approval
    ("Marketing","Marketing"), # Design Approval
    ("the HoD","the HoD"), # nan
    ("The retailer","The retailer"), # nan
    ("the expert","the expert"), # Hospital Workflow
    ("the referee","the referee"), # Hospital Workflow
    ]

#check for articles, plurality, upper and lower case, abbreviations, etc.
small_deviations = [
    ("A customer","the customer"), # Computer Repair Service
    ("computer repair service (CRS)","CRS"), # Computer Repair Service
    ("an employee","the employee"), # Expense Handling Process
    ("a manager","the manager"), # Expense Handling Process
    ("a customer","the customers"), 
    ("The payment method used in the billing details","the payment method"), 
    ("Swapfiets store employees","store employees"), # Bicycle Rental Process
    ("the customer","customer"), # Bicycle Rental Process
    ("a seller","the seller"), # Long term renting process
    ("sellers","the seller"), # Long term renting process
    ("The GRM","the Global Relationship Manager (GRM)"), # nan
    ("a Customer Service Consultant (CSC)","the CSC"), # nan
    ("the expert","an Expert"), # Hospital Workflow
    ("Referees","referee"), # Hospital Workflow
    ]

different_roles_same_process = [
    ("A customer","CRS"), # Computer Repair Service
    ("customer","computer repair service(CRS)"), # Computer Repair Service
    ("an employee","the manager"), # Expense Handling Process
    ("a manager","finance"), # Expense Handling Process
    ("storage facility used by Puori","Puori"), # Quality Control of medical products
    ("Ellipse","Eurofarms"), # Quality Control of medical products
    ("a customer","The payment method used in the billing details"), # Mikkeller webshop
    ("Swapfiets online customer service","Swapfiets store employees"), # Bicycle Rental Process
    ("the staff","Swapfiets store employees"), # Bicycle Rental Process
    ("the seller","the employee"), # Long term renting process
    ("management","the revenue"), # Long term renting process
    ("P&R","Marketing"), # Design Approval
    ("Marketing","Product Owner"), # Design Approval
    ("the CSC","the HoD"), # nan
    ("The retailer","the Finance department"), # nan
    ("the director","the manager"), # Hospital Workflow
    ("the manager","an expert"), # Hospital Workflow
    ]

def get_similarity(pairs, group):
    result = []
    for pair in pairs:
        embedding1 = model.encode(pair[0], convert_to_tensor=True)
        embedding2 = model.encode(pair[1], convert_to_tensor=True)
        similarity = model.similarity(embedding1, embedding2) #HERE CHECK FOR SIMILARITY
        result.append(round(float(similarity),4))
    print("Results for", group, ":")
    print("Average:", round(np.average(result),4))
    print("Minimum:", round(min(result),4))
    print("Maximum:", round(max(result),4), "\n")
    print(result, "\n")

get_similarity(same_roles, "same roles")
get_similarity(small_deviations, "small deviations")
get_similarity(different_roles_same_process, "different roles within the same process")
```
